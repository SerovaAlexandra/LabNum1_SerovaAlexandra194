Следующий код прочитает первичный файл и занесет все элементы файла (включая пробелы, знаки табуляции, знаки препинания) в список, в отдельный список занесет слова.  Затем прочитывается второй файл (функция file_first_maker берет на вход имя второго файла), список правок в этом файле мы будем хранить в переменной li_word, список элементов исходного файла+список правок - в переменной first_li

```python
def file_first_maker(filename):
    change_li=[]
    with open (filename, encoding='utf-8') as ff:  #читаем файл, в котором произвели какие-то правки по сравнению с файлом "first_file.txt"
        text1=ff.read()
        sec_li=[]
        for item in text1: #создаем список элементов строки, записанной в файле
            sec_li.append(item)
        list_of_tokens1=text1.split()  #отдельно создаем список токенов (в таком случае при сопоставлении с первичным файлом изменение формы слова будет считаться добавлением нового слова)
        tabs_counter1=sec_li.count(' ')
        enter_counter1=sec_li.count('\n')
        prep_counter1=sec_li.count(',')+sec_li.count('.')+sec_li.count(':')+sec_li.count('-')+sec_li.count('""')+sec_li.count(';')+sec_li.count('?')+sec_li.count('!')+sec_li.count('—')
    with open ('first_file.txt', encoding='utf-8') as f: #читаем исходный файл, чтобы сопоставить его со вторым
        text = f.read()
        first_li=[]
        for elem in text:
            first_li.append(elem) 
    tabs_counter=first_li.count(' ')
    enter_counter=first_li.count('\n')
    prep_counter=first_li.count(',')+first_li.count('.')+first_li.count(':')+first_li.count('-')+first_li.count('""')+first_li.count(';')+first_li.count('?')+first_li.count('!')+first_li.count('—')
    list_of_tokens=text.split() #также отдельно создаем список токенов в этом файле
    li_word=[] #список, в котором в виде строковых переменных (для слов) или вложенных списков (для пробелов, знаков препинания, переносов строки) будут храниться правки
    if len(list_of_tokens1) != len(list_of_tokens):
        for word in list_of_tokens1:   #проверяем, не было ли добавления слова
            if word in list_of_tokens:
                pass
            else:
                li_word.append(word) #добавляем дописанное в файл слово с список с правками
        for sm_word in list_of_tokens: #проверяем, не было ли удаления слова
            if sm_word in list_of_tokens1:
                pass
            else:
                th='minus_word'
                li_word.append(th) #отмечаем в списке правок удаление слова по сравнению с первоначальным документом
    if tabs_counter1<tabs_counter:  #если убрали пробел, добавляем информацию об этом в список+количество удалений пробелов (h)
        h=tabs_counter-tabs_counter1
        x=['minus_tab']
        x.append(h)
        li_word.append(x)
    elif tabs_counter1>tabs_counter: #то же, но для оперции добавления пробела
        g=tabs_counter1-tabs_counter
        z=[' ']
        z.append(g)
        li_word.append(z)
    else:  #если количество пробелов не изменилось, оставляем все как есть
        pass
    if enter_counter1<enter_counter:   #если убрали перенос, добавляем информацию об этом в список+количество удалений переносов (k)
        k=enter_counter-enter_counter1
        y=['minus_enter']
        y.append(k)
        li_word.append(y)
    elif enter_counter1>enter_counter: #то же, но для оперции добавления переноса
        l=enter_counter1-enter_counter
        m=['\n']
        m.append(l)
        li_word.append(m)
    else:
        pass
    if prep_counter1<prep_counter:  #если убрали знак препинания, добавляем информацию об этом в список+количество удалений знаков препинания (o)
        o=prep_counter-prep_counter1
        p=['minus_prep']
        p.append(o)
        li_word.append(p)
    elif prep_counter1>prep_counter: #для операции добавления знака препинания
        u=prep_counter1-prep_counter
        t='prep' + u + 'times'
        li_word.append(t)
    else:
        pass
    first_li.append(li_word) #список, описывающий содержимое измененного файла: список знаков исходного текстового файла+список правок
    with open ('file_changes.txt', 'w', encoding='utf-8') as ch:
        ch.write(str(first_li.pop()))
    with open ('new_file_like_first.txt', 'w', encoding='utf-8') as gh:
        first_li.pop()
        s=''.join(first_li)   #преобразуем список в строку и записываем в файл - в нем будет восстановлена первая версия текстового файла
        gh.write(s)
        
        
file_first_maker('second_smth.txt') #в файле добавлен пробел, перенос строки, слово "многолюдную"   
```
